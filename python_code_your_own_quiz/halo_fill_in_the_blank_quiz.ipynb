```python
'''
Steps:
- Greet player and ask for difficulty of quiz
    - Prompt user again for invalid difficulties
- Print full quiz
- Split up quiz
- Search along split quiz for a blank
- Prompt user for an answer:
    - If answer is correct: replace blank with correct answer and print quiz
    - Else: prompt user again
- Continue to next blank
- If complete congratulate player
'''


from random import randint

##### GLOBAL VARIABLES
spartan_number = str(randint(100,999))
spartan = 'Spartan-' + spartan_number

filling = ['___1___', '___2___', '___3___', '___4___']

difficulty_easy = ("\nEasy difficulty - please provide a noun to fill in the corresponding blank (case-sensitive):\n" +
                    "\n1. You heard the ___1___. Move like you've got a purpose!" +
                    "\n2. Get up, SO I CAN ___2___ YOU AGAIN!" +
                    "\n3. I'm proud of you, ___3___." +
                    "\n4. ___4___ said all these quotes."
                  )
easy_answers = ['lady', 'KILL', 'marines', 'Avery Johnson']

difficulty_normal = ("\nNormal difficulty - please provide a noun to fill in the corresponding blank (case-sensitive):\n" +
                    "\n1. Let's stay focused. Halo...how do we use it against the ___1___." +
                    "\n2. Yes...activate Halo's defenses, and destroy the ___2___, which is why we brought the Index to the Control Center." +
                    "\n3. This is ___3___, can anyone hear me? Over." +
                    "\n4. ___4___ said all these quotes."
                  )
normal_answers = ['Covenant', 'flood', 'Spartan-117', 'John-117']

difficulty_heroic = ("\nHeroic difficulty - please provide a noun or number to fill in the corresponding blank (case-sensitive):\n" +
                    "\n1. I'm reading thousands of types of flora. The ___1___ ecosystem is incredibly sophisticated." +
                    "\n2. New traffic on the Covenant Battlenet! I've found Captain Keyes! He's being held on a Covenant cruiser, the ___2___. (hint: three words)" +
                    "\n3. Cortana to Echo ___3___! Two Covenant banshees are approaching on your six! Evade, say again, evade!" +
                    "\n4. ___4___ said all these quotes."
                  )
heroic_answers = ['ring\'s', 'Truth and Reconciliation', '419', 'Cortana']

difficulty_legendary = ("Legendary difficulty - please provide a noun or number to fill in the corresponding blank (case-sensitive):\n" +
                    "\n1. We must collect the ___1___ before we can activate the installation." +
                    "\n2. Unfortunately, my usefulness to this particular endeavor has come to an end. Protocol does not allow units with my classification to perform a task as important as the reunification of the index with the Core. That final step is reserved for you, ___2___." +
                    "\n3. Alas, activating ___3___ will end all sentient life in the galaxy." +
                    "\n4. ___4___ said all these quotes."
                  )
legendary_answers = ['Index', 'Reclaimer', 'Halo', '343 Guilty Spark']



##### HELPER FUNCTIONS

def difficulty_select(difficulty):
    """
    Difficulty selection of Fill-in-the-Blank Quiz
    Input: difficulty (string) chosen by player
    Output: returns quiz of corresponding difficulty
    """
    if difficulty == 'easy':
        return play(difficulty_easy, easy_answers)
    if difficulty == 'normal':
        return play(difficulty_normal, normal_answers)
    if difficulty == 'heroic':
        return play(difficulty_heroic, heroic_answers)
    if difficulty == 'legendary':
        return play(difficulty_legendary, legendary_answers)
    else:
        difficulty = raw_input('Is that sass ' + spartan + '! Please indicate the difficulty you\'re most comfortable with: choose between Easy, Normal, Heroic, or Legendary -> ')
        difficulty_select(difficulty)


def blank_find(word, filling):
    """
    Checks to see if there is a blank in the quiz
    Input: takes a string (word) and a list (filling)
    Output: returns None or the element from filling
    """
    for blank in filling:
        if blank in word:
            return blank
    return None


def play(quiz, answers):
    """
    Playing the core game: takes answers, checks answers, replaces blanks with correct answers
    Input: quiz text (string) and quiz answers (list)
    Output: quiz progression and if you have successfully completed the quiz
    """
    print quiz
    answers_index = 0
    quiz_split = quiz.split()
    # checks through each element in split and prompts player for answer for fill in the blank
    for element in quiz_split:
        # Checks for a blank
        substitution = blank_find(element, filling)
        if substitution != None:
            # prompt player
            user_answer = raw_input('\nWhat is your answer for ' + substitution + ' -> ')
            while user_answer != answers[answers_index]:
                user_answer = raw_input('That answer is incorrect. A grunt could do better than that. Please submit another answer for ' + substitution + ' -> ')
            # replaces a blank with the correct answer
            if user_answer == answers[answers_index]:
                print '\nThat is correct'
                quiz = quiz.replace(substitution, user_answer)
                print quiz
                answers_index += 1
    print "\nCONGRATULATIONS " + spartan + "!!! You are now ready to rid the universe of the Covenant and Flood!!!"

##### GAME SETUP. Let's play!!!

def introduction():
    """
    Introduction to the game and initiates the game
    Input: players name ('string'), difficulty choice ('string')
    Output: starts the quiz
    """
    introduction = raw_input('Welcome Spartan. We have an assignment for you. First could I get your name: ')
    chit_chat = 'Thanks ' + introduction + ', we have assigned you as ' + spartan
    print chit_chat
    assignment = '\nWe are about to head out on a mission to save Earth from total destruction. You are humanities last hope. Before you head out on your mission. We will have to quiz you on your current understanding of the Halo-verse to better understand your skill level.'
    print assignment
    difficulty = raw_input('\nPlease indicate the difficulty you\'re most comfortable with: choose between Easy, Normal, Heroic, or Legendary -> ').lower()
    difficulty_select(difficulty)

introduction()
```
